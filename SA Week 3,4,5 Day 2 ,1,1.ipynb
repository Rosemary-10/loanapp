```python
# Importing train test split library
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split (x, y, test_size=0.2, random_state=20)
```

```python
# Checking the shape of my data
df.shape, df_encoded.shape, x_train.shape, x_test.shape
```

```python
# Importing the algorithm

from sklearn import svm   ##svm = support vector machine
from sklearn.linear_model import LogisticRegression

# Importing the accuracy model

from sklearn.metrics import accuracy_score
```

```python
# Using support Vector Model

classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(x_train, y_train)
```

### Model Evaluation

```python
# Training Data Accuracy

xtrain_pred = classifier.predict(x_train)
training_data_accuracy = accuracy_score(xtrain_pred, y_train)
```

```python
print('Accuracy Score', training_data_accuracy)
```

```python
# Testing Data Accuracy

xtest_pred = classifier.predict(x_test)
testing_data_accuracy = accuracy_score(xtest_pred, y_test)
```

```python
print('Accuracy Score', testing_data_accuracy)
```
